```python
import pandas as pd

data1 = {
    'Número da Nota Fiscal': [1001, 1002, 1003, 1004, 1005],
    'Valor Total': [150.50, 300.00, 75.25, 500.80, 1000.00],
    'Emissor': ['Empresa A', 'Empresa B', 'Empresa A', 'Empresa C', 'Empresa B'],
    'Destinatário': ['Cliente 1', 'Cliente 2', 'Cliente 3', 'Cliente 4', 'Cliente 5'],
    'Quantidade': [10, 1, 5, 2, 3],
    'Impostos': [22.58, 45.00, 11.27, 75.12, 150.00],
    'Data da Auditoria': ['2023-06-10', '2023-07-15', '2023-08-20', '2023-09-25', '2023-10-30'],
    'Resultado da Auditoria': ['Aprovada', 'Aprovada', 'Reprovada', 'Aprovada', 'Reprovada'],
    'Data de Regularização': ['2023-06-12', '2023-10-10', '2023-08-25', '2023-09-26', '2023-11-05'],
    'Status de Regularização': ['', 'Regularizada', 'Regularizada', '', 'Pendente']
}

data2 = {
    'Número NF': [1001, 1002, 1004, 1003, 1006],
    'Data de Emissão': ['2023-01-10', '2023-02-15', '2023-03-20', '2023-04-25', '2023-05-30'],
    'Descrição dos Produtos/Serviços': ['Produto X', 'Serviço Y', 'Produto Z', 'Serviço X', 'Produto Y'],
    'Valor Unitário': [15.05, 300.00, 15.05, 250.40, 333.33],
    'Observações da Auditoria': ['', 'Produtos com valores corretos', 'Produtos com valores corretos', '', 'Produtos danificados'],
    'Ações Corretivas': ['', '', 'Correção dos valores dos produtos', '', 'Aumento da quantidade'],
}

# Criando o DataFrame
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# Exercício 1: Realize um merge entre os dois DataFrames baseado no Número da Nota Fiscal. 
# Quantos registros resultam do merge?
pd.merge(df1, df2, left_on = 'Número da Nota Fiscal', right_on='Número NF')
```

```python
# Exercício 2: Combine os DataFrames e selecione apenas as colunas: "Data de Emissão", "Emissor", "Destinatário", "Valor Total".
df1.merge(df2, left_on = 'Número da Nota Fiscal', right_on='Número NF')[['Data de Emissão', 'Emissor','Destinatário', 'Valor Total']]
```

```python
# Exercício 3: Calcule o total de vendas realizadas por cada emissor onde o Valor da NF é superior a 200. 
df_merge = pd.merge(df1, df2, left_on = 'Número da Nota Fiscal', right_on='Número NF', how='left')
df_merge
```

```python
df_merge[df_merge['Valor Total'] > 200]['Emissor'].value_counts()
```

```python
# Exercício 4: Quais NF possuem o Valor Total divergente do Valor Unitário * Quantidade?
df_merge = df1.merge(df2, left_on = 'Número da Nota Fiscal', right_on='Número NF')
```

```python
df_merge['Valor Real'] = df_merge.apply(lambda x: x['Valor Unitário'] * x['Quantidade'], axis=1)
```

```python
df_merge[df_merge['Valor Total'] != df_merge['Valor Real']]
```

```python
# Exercício 5: Crie uma nova coluna chamada "Diferença entre Datas" que calcule a diferença em dias entre "Data da Auditoria" e 
# "Data de Regularização" somente das notas presentes em ambas os DataFrames.
df_merge = df1.merge(df2, left_on = 'Número da Nota Fiscal', right_on='Número NF')
```

```python
df_merge['Data da Auditoria'] = pd.to_datetime(df_merge['Data da Auditoria'])
df_merge['Data de Regularização'] = pd.to_datetime(df_merge['Data de Regularização'])
```

```python
df_merge['Data de Regularização'] - df_merge['Data da Auditoria']
```
